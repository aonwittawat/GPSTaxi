```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None 
```

```python
# Put dataset on my github repo
df = pd.read_csv('https://raw.githubusercontent.com/aonwittawat/GPSTaxi/master/2019-08-01.csv')
```

```python
# Exploratory Data Analysis
df['hour'] = pd.to_datetime(df['timestamp']).dt.hour
df.head(5)
```

```python
# https://seaborn.pydata.org/generated/seaborn.boxplot.html
# Changing default seaborn/matplotlib to be more readable
plt.figure(figsize=(15, 10))
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

# Look at Boxplot with Seaborn
ax = sns.boxplot(x='hour', y='location', data=df, color="skyblue")
ax.set_title('Sample GPS Data from Taxi', fontsize = 25)
ax.set_ylabel('Distance (m.)', fontsize = 20)
ax.set_xlabel('Hour', fontsize = 20)

# Show the datapoints on top of the boxes:
#sns.swarmplot(x="hour", y="location", data=df, color=".05")

# Save boxplot as image
# plt.savefig('sample_output.png')
```
